```python
import pandas as pd
```

```python
csvfile = "IMDB-Movie-Data.csv"
movie_data = pd.read_csv(csvfile)
movie_data
```

```python
melt = pd.melt(movie_data, id_vars=["Title", "Genre", "Description", "Director", "Year", "Rating", "Revenue (Millions)"], value_vars=['Actors', "Unnamed: 5", "Unnamed: 6", "Unnamed: 7"], value_name="Actor_Name")
melt
```

```python
melt_pd = melt[["Actor_Name", "Title", "Genre", "Description", "Director", "Year", "Rating", "Revenue (Millions)"]]
melt_pd
```

```python
melted = melt_pd.sort_values("Title")
melted
```

```python
melted_pd = melted.reset_index()
melted_pd.head()
```

```python
melted__pd = melted_pd.drop(["index"], axis=1)
melted__pd.head()
```

```python
melted__pd.to_csv("IMDBACTOR.csv", index=False, header=True)
```

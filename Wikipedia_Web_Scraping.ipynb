```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
import time
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_people_from_North_Carolina'
browser.visit(url)
```

```python
html = browser.html
# Parse HTML with Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')

actors = soup.find_all('div',class_='div-col columns column-width')

a_to_l = actors[2].find_all('a')
m_to_z = actors[3].find_all('a')


actors_list=[]

info_a_to_l =actors[2].find_all('li')
info_m_to_z = actors[3].find_all('li')
for items in info_a_to_l:
    actors_list.append(items.text)
for items in info_m_to_z:
    actors_list.append(items.text)
```

```python
actors_db = pd.DataFrame(actors_list)
actors_db.columns=["name"]
```

```python
actors_db = actors_db[actors_db.name.str.contains("act")]
actors_db.head(5)
```

```python
actors = actors_db['name'].str.split(",",  expand = True)
actor_name = actors[0]
```

```python
actor_name = pd.DataFrame(actor_name)
actor_name = pd.DataFrame(actor_name[0].str.replace(r"\s\(.*\)",""))
actor_name.columns=["Name"]
```

```python
actor_name.head()
```

```python
bio = pd.DataFrame(actors_db['name'].str.split('\),').str[-1])
bio = pd.DataFrame(bio['name'].str.split('\(').str[-1])
bio = pd.DataFrame(bio['name'].str.replace('\)',""))
bio = pd.DataFrame(bio['name'].str.replace('actress',""))
```

```python
birth_city = bio
birth_city.head()
```

```python
actor_info_db =actor_name.join(birth_city)
actor_info_db=actor_info_db.reset_index(drop=True)
```

```python
actor_info_db=actor_info_db.rename(columns={"Name":"Actor Name", "name": "City of Birth"})
actor_info_db.head(70)
```

```python
for index, row in actor_info_db.iterrows():
     print (index,row['Actor Name'],row["City of Birth"])
```
